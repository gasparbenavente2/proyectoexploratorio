```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
def get_url(i, j):
    return f'https://cl.linkedin.com/jobs/data-scientist-empleos?position={j}&pageNum={i}'
```

```python
i = 0
j = 1
c = 0

htmls = []

while True:
    while True:
        url = get_url(i, j)
        #html = requests.get(url).content
        #soup = bs(html)

        htmls.append(url)

        j += 1
        if j == 26:
            j = 1
            break
    i += 1

    c += 1
    if c == 39:
        break

for i in htmls:
    print(i)
```

```python
c = 0
for url in htmls:
    html = requests.get(url).content
    soup = bs(html)

    title = soup.select_one('body > div.base-serp-page > div > section > div.details-pane__content.details-pane__content--show > section > div > div.top-card-layout__entity-info-container.flex.flex-wrap.papabear\:flex-nowrap > div > a > h2')
    print(title)
    c += 1
    if c == 10:
        break
```
